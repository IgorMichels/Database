```python
import pandas as pd
import numpy as np

local = 'C:/Users/igorm/Desktop/FGV/4ยบ Semestre/07 - Banco de Dados/Csvs/'
novo_local = 'C:/Users/igorm/Desktop/FGV/4ยบ Semestre/07 - Banco de Dados/'
arquivos = ['2019-01', '2019-02', '2019-03', '2019-04',
            '2019-05', '2019-06', '2019-07', '2019-08',
            '2019-09', '2019-10', '2019-11', '2019-12',
            '2020-01', '2020-02', '2020-03', '2020-04',
            '2020-05'] # lista de arquivos que temos sem o ' compressed.csv'
```

```python
dataframe = pd.read_csv(local + arquivos[6] + ' compressed.csv', delimiter = ';')
```

```python
order = [1, 2, 11, 12, 13, 14, 15, 16, 17, 18, 3, 4, 5, 6, 7, 8, 9, 10]
dataframe = dataframe[[dataframe.columns[i] for i in order]]
dataframe
```

```python
delay = dataframe.iloc[:, :10]
delay
```

```python
column_names = ["delay_cause"]
delay_cause = pd.DataFrame(columns = column_names)

line = 0
for i in delay['DEP_DELAY']:
    if i <= 0 or np.isnan(i):
        new_row = {'delay_cause' : ''}
    else:
        if np.isnan(delay['CARRIER_DELAY'][line]):
            maior = 0
            k = ""
        else:
            maior = delay['CARRIER_DELAY'][line]
            k = 1
        if np.isnan(delay['WEATHER_DELAY'][line]) == False and maior < delay['WEATHER_DELAY'][line]:
            k = 2
            maior = delay['WEATHER_DELAY'][line]
        if np.isnan(delay['NAS_DELAY'][line]) == False and maior < delay['NAS_DELAY'][line]:
            k = 3
            maior = delay['NAS_DELAY'][line]
        if np.isnan(delay['SECURITY_DELAY'][line]) == False and maior < delay['SECURITY_DELAY'][line]:
            k = 4
            maior = delay['SECURITY_DELAY'][line]
        if np.isnan(delay['LATE_AIRCRAFT_DELAY'][line]) == False and maior < delay['LATE_AIRCRAFT_DELAY'][line]:
            k = 5
            maior = delay['LATE_AIRCRAFT_DELAY'][line]
        new_row = {'delay_cause' : k}
    line += 1
    if line % 1000 == 0:
        print(line)
    delay_cause = delay_cause.append(new_row, ignore_index = True)
    
delay_cause
```

```python
frames = [dataframe, delay_cause]
df = pd.concat(frames, axis = 1)
df
```

```python
df.drop('ORIGIN_CITY_NAME', 1, inplace = True)
df.drop('ORIGIN_STATE_ABR', 1, inplace = True)
df.drop('ORIGIN_STATE_NM', 1, inplace = True)
df.drop('DEST_CITY_NAME', 1, inplace = True)
df.drop('DEST_STATE_ABR', 1, inplace = True)
df.drop('DEST_STATE_NM', 1, inplace = True)
df.drop('LATE_AIRCRAFT_DELAY', 1, inplace = True)
df.drop('WEATHER_DELAY', 1, inplace = True)
df.drop('NAS_DELAY', 1, inplace = True)
df.drop('SECURITY_DELAY', 1, inplace = True)
df.drop('CARRIER_DELAY', 1, inplace = True)
df.to_csv(local + '2019-07 - final.csv', sep = ';')
```
