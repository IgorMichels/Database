```python
import pandas as pd

local = 'C:/Users/igorm/Desktop/FGV/4ยบ Semestre/07 - Banco de Dados/Csvs/'
arquivos = ['2019-01', '2019-02', '2019-03', '2019-04',
            '2019-05', '2019-06', '2019-07', '2019-08',
            '2019-09', '2019-10', '2019-11', '2019-12',
            '2020-01', '2020-02', '2020-03', '2020-04',
            '2020-05'] # lista de arquivos que temos sem o ' compressed.csv'
```

```python
# juntando todos os dados num sรณ data frame

dataframe = pd.read_csv(local + arquivos[0] + ' compressed.csv', delimiter = ';')

for arquivo in arquivos:
    if arquivo == "2019-01":
        pass
    else:
        dataframe2 = pd.read_csv(local + arquivo + ' compressed.csv', delimiter = ';')
        frames = [dataframe, dataframe2]
        dataframe = pd.concat(frames)
        
```
